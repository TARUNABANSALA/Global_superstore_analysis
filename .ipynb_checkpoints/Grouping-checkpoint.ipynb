```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import ticker as mtick
import seaborn as sns
```

```python
# import datasets
stores = pd.read_csv('store.csv',  encoding = 'unicode_escape', engine ='python')
```

```python
# data head
stores.head(2)
```

```python
# Remove column 'row_id'
stores = stores.drop('row_id', axis=1)
```

```python
stores.head(2)
```

```python
grouped = stores.groupby(['customer_id', 'order_id', 'product_id'])
```

```python
# You can perform operations on the grouped data, like calculating the sum of sales for each group
sum_sales_per_group = grouped['sales'].sum().reset_index()
```
